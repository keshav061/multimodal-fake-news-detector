```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from nltk.tokenize import sent_tokenize
import re
```

```python
contextCountsFake = []
contextCountsTrue = []

wordCountsFake = []
wordCountsTrue = []

for i in range(len(NewsISOT)):
    article = NewsISOT.loc[i,'text']

    if pd.isna(article):
        continue

    contexts = 0
    
    for sentence in sent_tokenize(article):
        clauses = re.split(r'\s*[:;,]\s*',sentence)

        for clause in clauses:
            clause = clause.strip()

            if clause:
                contexts += 1

                if NewsISOT.loc[i,'authenticity'] == 0:
                    wordCountsFake.append(len(clause.split()))
                else:
                    wordCountsTrue.append(len(clause.split()))

    if NewsISOT.loc[i,'authenticity'] == 0:
        contextCountsFake.append(contexts)
    else:
        contextCountsTrue.append(contexts)
```

```python
for name, data in [
    ("Fake Contexts", contextCountsFake),
    ("True Contexts", contextCountsTrue),
    ("Fake Words", wordCountsFake),
    ("True Words", wordCountsTrue)
]:
    s = pd.Series(data)

    print(f"\n{name}")
    print(f"Mean        :{s.mean():.2f}")
    print(f"Median      :{s.median():.2f}")
    print(f"Mode        :{list(s.mode().values)}")
    print(f"Std         :{s.std():.2f}")
    print(f"Variance    :{s.var():.2f}")
    print(f"Min         :{s.min()}")
    print(f"Max         :{s.max()}")
    print(f"25%         :{s.quantile(0.25)}")
    print(f"75%         :{s.quantile(0.75)}")
    print(f"IQR         :{s.quantile(0.75)-s.quantile(0.25)}")
    print(f"Skewness    :{s.skew():.2f}")
    print(f"Kurtosis    :{s.kurt():.2f}")
```

```python
trueContexts = [c for c in contextCountsTrue if 1 <= c <= 120]
fakeContexts = [c for c in contextCountsFake if 1 <= c <= 120]
trueWords = [c for c in wordCountsTrue if 1 <= c <= 60]
fakeWords = [c for c in wordCountsFake if 1 <= c <= 60]
```

```python
from collections import Counter
import pandas as pd

TrueLengthCounter = Counter()
FakeLengthCounter = Counter()

for i in range(len(NewsISOT)):
    article = NewsISOT.loc[i, "text"]

    if pd.isna(article):
        continue

    words = article.split()

    if NewsISOT.loc[i, "authenticity"] == 0:
        for word in words:
            FakeLengthCounter[len(word)] += 1
    else:
        for word in words:
            TrueLengthCounter[len(word)] += 1

true_df = pd.DataFrame({
    "Length": list(TrueLengthCounter.keys()),
    "Frequency": list(TrueLengthCounter.values())
}).sort_values("Length")

fake_df = pd.DataFrame({
    "Length": list(FakeLengthCounter.keys()),
    "Frequency": list(FakeLengthCounter.values())
}).sort_values("Length")
```

```python
true_total = true_df["Frequency"].sum()
fake_total = fake_df["Frequency"].sum()

true_mean = (true_df["Length"] * true_df["Frequency"]).sum() / true_total
fake_mean = (fake_df["Length"] * fake_df["Frequency"]).sum() / fake_total

print("TRUE NEWS")
print(f"Total Words : {true_total:,}")
print(f"Unique Lengths : {len(true_df)}")
print(f"Mean Length : {true_mean:.3f}")

print()

print("FAKE NEWS")
print(f"Total Words : {fake_total:,}")
print(f"Unique Lengths : {len(fake_df)}")
print(f"Mean Length : {fake_mean:.3f}")
```

```python
from collections import Counter

# Top 10000 counters
TopTrue = Counter(dict(TrueCounter.most_common(20000)))
TopFake = Counter(dict(FakeCounter.most_common(10000)))

# Union with summed frequencies
TopCombined = TopTrue + TopFake
```

```python
print(len(TopTrue))      # 20000
print(len(TopFake))      # 10000
print(len(TopCombined))  # Between 10000 and 30000
```

```python
selected_tokens = 0
total_tokens = sum(TrueCounter.values()) + sum(FakeCounter.values())

for word in TopCombined:
    selected_tokens += TrueCounter.get(word, 0)
    selected_tokens += FakeCounter.get(word, 0)

coverage = selected_tokens / total_tokens

print(f"Corpus Coverage: {coverage:.2%}")
```

```python
from collections import Counter

# Top words
TopTrueWords = set(word for word, _ in TrueCounter.most_common(20000))
TopFakeWords = set(word for word, _ in FakeCounter.most_common(10000))

# Union of vocabularies
CombinedWords = TopTrueWords | TopFakeWords

print(f"Top True words      : {len(TopTrueWords):,}")
print(f"Top Fake words      : {len(TopFakeWords):,}")
print(f"Combined vocabulary : {len(CombinedWords):,}")

# Total tokens in each corpus
true_total_tokens = sum(TrueCounter.values())
fake_total_tokens = sum(FakeCounter.values())
total_tokens = true_total_tokens + fake_total_tokens

# Tokens covered by the combined vocabulary
covered_true = sum(TrueCounter.get(word, 0) for word in CombinedWords)
covered_fake = sum(FakeCounter.get(word, 0) for word in CombinedWords)

covered_total = covered_true + covered_fake

# Coverage
true_coverage = covered_true / true_total_tokens
fake_coverage = covered_fake / fake_total_tokens
overall_coverage = covered_total / total_tokens

print(f"True Coverage    : {true_coverage:.2%}")
print(f"Fake Coverage    : {fake_coverage:.2%}")
print(f"Overall Coverage : {overall_coverage:.2%}")
```

```python
from collections import Counter
import pandas as pd
import string

# -----------------------------
# Preprocessing function
# -----------------------------
def preprocess_word(word):
    # Convert to lowercase
    word = word.lower()

    # Remove leading and trailing punctuation
    word = word.strip(string.punctuation)

    # Skip empty strings
    if not word:
        return None

    return word

# -----------------------------
# Build vocabularies
# -----------------------------
FakeCounter = Counter()
TrueCounter = Counter()

for _, row in NewsISOT.iterrows():

    text = row["text"]

    if pd.isna(text):
        continue

    words = []

    for word in text.split():

        word = preprocess_word(word)

        if word is not None:
            words.append(word)

    if row["authenticity"] == 0:
        FakeCounter.update(words)
    else:
        TrueCounter.update(words)

print("Fake vocabulary size :", len(FakeCounter))
print("True vocabulary size :", len(TrueCounter))
```

```python
import numpy as np
import pandas as pd

checkpoints = [100, 250, 500, 1000, 1250, 1500, 1750,
               2000, 2250, 2500, 2750, 3000,
               4000, 5000, 10000]

def coverage_statistics(counter):

    total_tokens = sum(counter.values())

    freq = np.array([c for _, c in counter.most_common(10000)])

    cum = np.cumsum(freq) / total_tokens

    return [cum[k-1] for k in checkpoints]


fake_cov = coverage_statistics(FakeCounter)
true_cov = coverage_statistics(TrueCounter)

coverage = pd.DataFrame({
    "Top Words": checkpoints,
    "Fake Coverage (%)": np.round(np.array(fake_cov)*100,2),
    "True Coverage (%)": np.round(np.array(true_cov)*100,2)
})

display(coverage)
```

```python
import matplotlib.pyplot as plt

def vocabulary_growth(counter, step=1000):

    words = list(counter.elements())

    vocab = set()

    x = []
    y = []

    for i, word in enumerate(words, 1):
        vocab.add(word)

        if i % step == 0:
            x.append(i)
            y.append(len(vocab))

    return x, y

xf, yf = vocabulary_growth(FakeCounter)
xt, yt = vocabulary_growth(TrueCounter)

plt.figure(figsize=(10,5))
plt.plot(xf, yf, label="Fake")
plt.plot(xt, yt, label="True")
plt.xlabel("Tokens Seen")
plt.ylabel("Vocabulary Size")
plt.legend()
plt.grid()
plt.show()
```

```python
def coverage(counter):

    freq = np.array([c for _, c in counter.most_common()])
    return np.cumsum(freq)/freq.sum()

fake_cov = coverage(FakeCounter)
true_cov = coverage(TrueCounter)

plt.figure(figsize=(10,5))

plt.plot(fake_cov,label="Fake")
plt.plot(true_cov,label="True")

plt.xlim(0,30000)

plt.xlabel("Top k Words")
plt.ylabel("Coverage")

plt.legend()
plt.grid()

plt.show()
```

```python
def summarize(counter):

    total_vocab=len(counter)

    total_tokens=sum(counter.values())

    top=counter.most_common(10000)

    top_vocab=len(top)

    top_tokens=sum(c for _,c in top)

    print("Top Vocabulary :",top_vocab)
    print("Tail Vocabulary :",total_vocab-top_vocab)

    print()

    print("Top Tokens :",top_tokens)
    print("Tail Tokens :",total_tokens-top_tokens)

    print()

    print("Tail Vocabulary % :",100*(total_vocab-top_vocab)/total_vocab)

    print("Tail Token % :",100*(total_tokens-top_tokens)/total_tokens)

print("Fake")
summarize(FakeCounter)

print()

print("True")
summarize(TrueCounter)
```
