```python
# Dropping the duplicate rows 

trueNews = trueNews.drop_duplicates().reset_index(drop=True)
fakeNews = fakeNews.drop_duplicates().reset_index(drop=True)

print(f'Shape of trueNews - {trueNews.shape}')
print(f'Shape of fakeNews - {fakeNews.shape}')
```

```python
# Dropping the exact title and text

trueNews = trueNews.drop_duplicates(subset=['title','text']).reset_index(drop=True)
fakeNews = fakeNews.drop_duplicates(subset=['title','text']).reset_index(drop=True)
```

```python
trueNews = trueNews[['title','text']]
fakeNews = fakeNews[['title','text']]
```

```python
# Dropping the exact title as the text are different due to typo errors in trueNews
# Retain the exact text as they are less and tell different styles of title in trueNews

trueNews = trueNews.drop_duplicates(subset=['title']).reset_index(drop=True)
```

```python
# Many rows of the fakeNews contains empty text so removing them

print((fakeNews["text"].str.strip() == "").sum())
fakeNews = fakeNews[fakeNews["text"].str.strip() != ""].reset_index(drop=True)
```

```python
# Dropping the rows with duplicate title or duplicate text as the other has just typo errors

print(f"No of duplicate titles - {fakeNews['title'].duplicated().sum()}")
print(f"No of duplicate texts - {fakeNews['text'].duplicated().sum()}")

fakeNews = fakeNews.drop_duplicates(subset=['title']).reset_index(drop=True)
fakeNews = fakeNews.drop_duplicates(subset=['text']).reset_index(drop=True)
```

```python
# Some rows in fakeNews contain just a single word which is mostly a link, so removing them

print(f"No of single word texts - {(fakeNews['text'].str.split().str.len() <= 1).sum()}")
print(f"No of single word titles - {(fakeNews['title'].str.split().str.len() <= 1).sum()}")

fakeNews = fakeNews[(fakeNews["text"].str.split().str.len() > 1) & (fakeNews["title"].str.split().str.len() > 1)].reset_index(drop=True)
```

### Almost all the text in trueNews begins with "CITY_NAME (Reuters) - " . So I will remove all these starting phrases.

```python
trueNews['text'] = (trueNews['text'].str.split('-',n=1).str[-1].str.lstrip())
```

```python
trueNews['text'].head(20)
```

### Most of the fakeNews data contain image / video sources in the end in their text which in uncommon in trueNews. So I will remove them. 

```python
fakeNewsText = fakeNews['text']
fakeNewsText.info()
```

```python
for i in range(len(fakeNewsText)):
    if (fakeNewsText[i])[-1] == '.':
        fakeNewsText[i] = (fakeNewsText[i])[:-1]
```

```python
import re

for i in range(50):
    article = fakeNewsText[i]
    parts = re.split(r'[?.!]',article)
    parts = [p.strip() for p in parts if p.strip()]
    print(parts[-1])
```

```python
# Removing the sources at last in the fakeNewsText

for i in range(len(fakeNewsText)):
    article = fakeNewsText[i]
    article = re.sub(r'[?.!][^?.!]*$','',article)
    article = article.strip()
    fakeNewsText[i] = article + '.'
```

```python
for i in range(50):
    article = fakeNewsText[i]
    parts = re.split(r'[?.!]',article)
    parts = [p.strip() for p in parts if p.strip()]
    if parts:
        print(parts[-1])
```

### Many of the titles in fakeNews contain words like WATCH, VIDEO, TWEET which are enclosed in brackets. All the brackets words are just emphasising and are not present in trueNews. I will remove the words in bracket

```python
fakeNewsTitle = fakeNews['title']
fakeNewsTitle.info()
```

```python
for i in range(len(fakeNewsTitle)):
    title = fakeNewsTitle[i]
    newTitle = re.sub(r"\s?\([^)]","",title)
    fakeNewsTitle[i] = newTitle
    fakeNewsTitle[i].lstrip()
```

### The fakeNewsTitle contains many words in full capital like BREAKING, ANNOYED, ENOUGH and many more. These are not abbreviations. These are just added to emphasise, these emphasis are not present in trueNews and is a direct source to detect fake news. So I will modify these capital words.

### It is difficult to select these capital words in the whole fake news dataset, and then classify whether they are used for emphasis or are abbreviations. So I will use a trick. I will make a list of capital words in true news titles. They are mostly abbreviations. And then use this list to make all other capital words to lowercase in the fake news dataset.

### Another observation is that all the words in the title of fake news dataset start with capital letter. So I will first make all the words to start with lowercase (except abbreviations list) and then make the first letter of the title to uppercase.

```python
trueNewsTitle = trueNews['title']

abbreviations = []
names = []

for i in range(len(trueNewsTitle)):
    title = trueNewsTitle[i]
    words = title.split()
    abbreviations.extend([word for word in words if word.isupper()])


for i in range(len(trueNewsTitle)):
    title = trueNewsTitle[i]
    words = title.split()
    for word in words[1:]:
        if word[0].isupper() and not word.isupper():
            names.append(word)

abbreviations = list(set(abbreviations))
names = list(set(names))

print(f'No of abbreviations - {len(abbreviations)}')
print(f'No of names - {len(names)}')
```

```python
for i in range(len(fakeNewsTitle)):
    title = fakeNewsTitle[i]
    words = title.split()
    for j in range(len(words)):
        if words[j] == 'A' or (words[j] not in names and words[j] not in abbreviations):
            words[j] = words[j].lower()
            
    fakeNewsTitle[i] = " ".join(words)
```

```python
# Now making the starting letter to uppercase

for i in range(len(fakeNewsTitle)):
    title = fakeNewsTitle[i]
    if title:
        title = title[0].upper() + title[1:]
    fakeNewsTitle[i] = title
```

```python
for i in range(50):
    print(fakeNewsTitle[i])
```

Most of the noise in fakeNewsTitle is removed. Now I will modify some words by manual inspection which are difficult to select.

```python
# Listing all the words whose first letter is uppercase 
# I will change these to lowercase if its a common word where uppercase is not required

firstLetterCapital = []

for i in range(len(fakeNewsTitle)):
    title = fakeNewsTitle[i]
    words = title.split()
    for word in words[1:]:
        if word[0].isupper() and not word.isupper():
            firstLetterCapital.append(word)

firstLetterCapital = list(set(firstLetterCapital))
print(len(firstLetterCapital))
#print(firstLetterCapital)
```

```python
fakeNames = ['Research','Child','Now','See','Maybe','Is','In','At','Says']

for i in range(len(fakeNewsTitle)):
    title = fakeNewsTitle[i]
    words = title.split()
    for j in range(len(words)):
        if words[j] in fakeNames:
            words[j] = words[j].lower()
            
    fakeNewsTitle[i] = " ".join(words)
```

```python
fakeNewsTitleWords = []

for i in range(len(fakeNewsTitle)):
    title = fakeNewsTitle[i]
    words = title.split()
    fakeNewsTitleWords.extend(words)

fakeNewsTitleWords = sorted(list(set(fakeNewsTitleWords)))
print(len(fakeNewsTitleWords))
for i in range(50):
    print(fakeNewsTitleWords[i])
```

```python
trueNewsTitleWords = []

for i in range(len(trueNewsTitle)):
    title = trueNewsTitle[i]
    words = title.split()
    trueNewsTitleWords.extend(words)

trueNewsTitleWords = sorted(list(set(trueNewsTitleWords)))

print(len(trueNewsTitleWords))
for i in range(50):
    print(trueNewsTitleWords[i])
```

### Many titles in fake news dataset contain '#' before words which is not present in true news dataset. So, I will remove all the '#'.

```python
# Removing the '[video]', ')','Watch:' as many fakeNewsTitle have these in the end

for i in range(len(fakeNewsTitle)):
    title = fakeNewsTitle[i]
    
    title = (title
             .replace("Watch:", "")
             .replace("Breaking:", "")
             .replace("Busted:", "")
             .replace("[video]", "")
             .replace("#", "")
             .replace(")", ""))

    title = " ".join(title.split())
    fakeNewsTitle[i] = title
```

### Saving the modified true and fake datasets

I am labelling trueNews to 1 and fakeNews to 0.

```python
fakeNewsISOT = pd.DataFrame({
    "title" : fakeNewsTitle,
    "text": fakeNewsText,
    "authenticity": 0
        })

trueNewsText = trueNews['text']

trueNewsISOT = pd.DataFrame({
    "title" : trueNewsTitle,
    "text": trueNewsText,
    "authenticity" : 1
        })
```

```python
fakeNewsISOT.to_csv("../data/processed/fakeNewsISOT.csv",index=False)
trueNewsISOT.to_csv("../data/processed/trueNewsISOT.csv",index=False)
```

```python
combineNews = pd.concat([fakeNewsISOT,trueNewsISOT], ignore_index=True)
NewsISOT = combineNews.sample(frac=1,random_state=42).reset_index(drop=True)
```
