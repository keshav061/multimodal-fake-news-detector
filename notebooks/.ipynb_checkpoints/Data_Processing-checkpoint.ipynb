# DATA PROCESSING

### In this notebook I will perform various analysis on the data. Based on the conclusions, I will prepare the dataset for feeding into the neural net.

### First, I will analyse the clauses in the text. By clause, I mean a sentence seperated by any of the punctuation marks (.?!,;:-). I interpret them as facts stated by the article. Each clause is a fact about the title.

```python
from nltk.tokenize import sent_tokenize
import re
```

#### nltk is a python library to separate sentences based on punctuation marks (.?!) while preserving the abbreviations.

```python
contextCountsFake = []
contextCountsTrue = []

wordCountsFake = []
wordCountsTrue = []

for i in range(len(NewsISOT)):
    article = NewsISOT.loc[i,'text']

    contexts = 0

    for sentence in sent_tokenize(article):
        clauses = re.split(r'\s*[:;,]\s*', sentence)

        for clause in clauses:
            clause = clause.strip()

            if clause:
                contexts += 1

                if NewsISOT.loc[i,'authenticity'] == 0:
                    wordCountsFake.append(len(clause.split()))
                else:
                    wordCountsTrue.append(len(clause.split()))
                    
    if NewsISOT.loc[i,'authenticity'] == 0:
        contextCountsFake.append(contexts)
    else:
        contextCountsTrue.append(contexts)
```

```python
def get_stats(data):
    s = pd.Series(data)

    return {
    "Mean":  s.mean(),
    "Median":  s.median(),
    "Mode":  ", ".join(map(str, s.mode().tolist())),
    "Variance":  s.var(),
    "Min":  s.min(),
    "Max":  s.max(),
    "25%":  s.quantile(0.25),
    "75%":  s.quantile(0.75),
    "IQR":  s.quantile(0.75) - s.quantile(0.25)
    }

contextTable = pd.DataFrame({
    "Fake": get_stats(contextCountsFake),
    "True": get_stats(contextCountsTrue)
})

wordTable = pd.DataFrame({
    "Fake": get_stats(wordCountsFake),
    "True": get_stats(wordCountsTrue)
})

print("Context Statistics")
display(contextTable)

print("Word Statistics")
display(wordTable)
```

```python
trueContexts = [c for c in contextCountsTrue if 1 <= c <= 120]
fakeContexts = [c for c in contextCountsFake if 1 <= c <= 120]
trueWords = [c for c in wordCountsTrue if 1 <= c <= 60]
fakeWords = [c for c in wordCountsFake if 1 <= c <= 60]
```

```python
#Context plot

fig, ax = plt.subplots(1,2,figsize=(14,5), sharey=True)

#Fake 
sns.histplot(
    fakeContexts,
    bins = 120,
    stat="frequency",
    ax=ax[0]
)
ax[0].set_title("Fake News")
ax[0].set_xlabel("Contexts per article")
ax[0].set_ylabel("Frequency")

#True
sns.histplot(
    trueContexts,
    bins=120,
    stat="frequency",
    ax=ax[1]
)

ax[1].set_title("True News")
ax[1].set_xlabel("Contexts per article")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Words plot

fig, ax = plt.subplots(1,2,figsize=(14,5), sharey=True)

#Fake 
sns.histplot(
    fakeWords,
    bins = 60,
    stat="frequency",
    ax=ax[0]
)
ax[0].set_title("Fake News")
ax[0].set_xlabel("Words per context")
ax[0].set_ylabel("Frequency")

#True
sns.histplot(
    trueWords,
    bins=60,
    stat="frequency",
    ax=ax[1]
)

ax[1].set_title("True News")
ax[1].set_xlabel("Words per context")
ax[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### Initially I thought to create embedding for each context as they state facts. But, in this process, there can be almost infinte contexts as even a single letter can change context entirely and there would be no generalization.

### Now I will create embeddings for words. For this I will analyse the unique words and create a vocabulary.

```python
from collections import Counter

TrueCounterNoisy = Counter()
FakeCounterNoisy = Counter()

for i in range(len(NewsISOT)):
    article = NewsISOT.loc[i,'text']

    words = article.split()

    if NewsISOT.loc[i,'authenticity'] == 0:
        FakeCounterNoisy.update(words)
    else:
        TrueCounterNoisy.update(words)
```

```python
print(f'Words in True News text: {len(TrueCounterNoisy)}')
print(f'Words in Fake News text: {len(FakeCounterNoisy)}')

CombinedWordsNoisy = set(TrueCounterNoisy) | set(FakeCounterNoisy)
print(f'Words in all texts: {len(CombinedWordsNoisy)}')
```

#### The words in the articles are noisy. Some noises are 
1) Uppercase and lowercase words are treated differently
2) There are many words that have a punctuation at the end.
3) Many words are enclosed in comma, bracket.
4) Many words have a comma, colon at the end.
5) Many words have - between them combining them. They should be separated into individual words.

#### Now I will build the vocabulary by eliminating the noise as much as possible.

```python
# Pre-processing the words 

def preprocess_words(word):

    # Convert to lowercase
    word = word.lower()

    # Remove the punctuations, signs at the start or end of words
    word = re.sub(r'^[^a-z0-9]+|[^a-z0-9]+$', '', word)

    # Splitting the numeric, alphabetical and - separated parts
    return re.findall(r"[a-z']+|\d+",word)
    
```

```python
from tqdm.notebook import tqdm

FakeCounter = Counter()
TrueCounter = Counter()

for _, row in tqdm(NewsISOT.iterrows(), total=len(NewsISOT)):
    text = row['text']

    words = []

    for word in text.split():
        keys = preprocess_words(word)

        if keys:
            words.extend(keys)

    if row["authenticity"] == 0:
            FakeCounter.update(words)
    else:
            TrueCounter.update(words)
```

```python
print("Fake vocabulary size: ", len(FakeCounter))
print("True vocabulary size: ", len(TrueCounter))

CombinedWords = set(TrueCounter) | set(FakeCounter)
print(f'Words in all texts: {len(CombinedWords)}')
```

#### After removing the noise, the vocabulary size has reduced significantly. Now to build a vocabulary, I will select the most frequently occuring words.

```python
checkpoints = [100, 250, 500, 1000, 1500, 2000, 2500, 3000, 4000, 5000, 8000, 10000, 15000]

def coverage_statistics(counter):

    total_words = sum(counter.values())
    freq = np.array([ c for _, c in counter.most_common(15000)])
    cum = np.cumsum(freq) / total_words

    return [cum[k-1] for k in checkpoints]

fakeCov = coverage_statistics(FakeCounter)
trueCov = coverage_statistics(TrueCounter)

coverage = pd.DataFrame({
    "Top Words": checkpoints,
    "Fake Coverage (%)": np.round(np.array(fakeCov)*100,2),
    "True Coverage (%)": np.round(np.array(trueCov)*100,2)
})

display(coverage)
```

### Now I will create the vocabulary by combining top 12000 words of text and title. I will add certain other IDs also.

```python
TOP_K = 12000

fakeTopWords = {word for word, _ in FakeCounter.most_common(TOP_K)}
trueTopWords = {word for word, _ in TrueCounter.most_common(TOP_K)}

fakeTopWordsTitle = {word for word, _ in FakeCounterTitle.most_common(TOP_K)}
trueTopWordsTitle = {word for word, _ in TrueCounterTitle.most_common(TOP_K)}

Vocab_ISOT = (
    fakeTopWords |
    trueTopWords |
    fakeTopWordsTitle |
    trueTopWordsTitle
)

print(f"Vocabulary size: {len(Vocab_ISOT)}")
```

```python
CombinedCounter = (FakeCounter + TrueCounter + FakeCounterTitle + TrueCounterTitle)

VocabCounter = Counter({
        word: CombinedCounter[word]
        for word in Vocab_ISOT
})

SortedVocab = VocabCounter.most_common()
```

```python
SPECIAL_TOKENS = ["<PAD>", "<UNK>", "<SOS>", "<EOS>", "<NUM1>", "<NUM2>", "<NUM3>", "<NUM4>","<NUM4+>",
                  "<LINK>", "<#>", "<$>", "<%>", "<@>", "<SOURCE>" ]
```

```python
CombinedCounter = (FakeCounter + TrueCounter + FakeCounterTitle + TrueCounterTitle)

SortedVocab = [
    (word, CombinedCounter[word])
    for word in Vocab_ISOT
]

SortedVocab.sort(key=lambda x: x[1], reverse=True)

specialSet = set(SPECIAL_TOKENS)
SortedVocab = [(w, f) for w, f in SortedVocab if w not in specialSet]

Vocab = []

for token in SPECIAL_TOKENS:
    Vocab.append((token, 0))

Vocab.extend(SortedVocab)

word2id = {word: idx for idx, (word, _) in enumerate(Vocab)}
id2word = {idx: word for word, idx in word2id.items()}
```
